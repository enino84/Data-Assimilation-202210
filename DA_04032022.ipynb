Sampling Errors - $\frac{1}{\sqrt{N}}$, $N$ denotes sample size

$\psi^f = {\bf x}^b$, 
$\psi^a = {\bf x}^a$, and 
${\bf P}^f = {\bf B}$

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
N = 2;

#Forecast error distribution (background error distribution)
xf = np.array([1,1], dtype=np.float32);
Pf = np.array([[4, 2],[2,5]], dtype=np.float32);
```

```python
Xf = np.random.multivariate_normal(xf, Pf, 100);
```

```python
sns.kdeplot(x=Xf[:,0], y=Xf[:,1], fill=True, cmap='viridis')
```

```python
Pf_e = np.cov(Xf.T)
```

# Matriz de covarianza real

```python
sns.heatmap(Pf, annot=True, cmap='viridis')
```

# Matriz de covarianza estimada por medio de muestras

```python
sns.heatmap(Pf_e, annot=True, cmap='viridis')
```

```python
error = [];
for N in range(1000,100000,500):
  Xf = np.random.multivariate_normal(xf, Pf, N);
  Pf_e = np.cov(Xf.T)
  error_i = np.linalg.norm(Pf_e-Pf);
  error.append(error_i);
error = np.array(error, dtype=np.float32);
```

```python
plt.plot(np.arange(1000,100000,500), error)
plt.xlabel('$n$', fontdict={'size':18})
plt.ylabel('$\\||{\\bf P}^f-{\\bf P}^f_e\\||$', fontdict={'size':18})
```
